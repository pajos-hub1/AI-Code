Syntax and Data

```python
# syntax and data
# variables
name = "Tunde"
age = 35
height = 1.75
is_student = True

# variable naming rules
first_name = "Bola"
my_total_salary = 20000

# data types
print(type(name)) # (str)
print(type(age)) #(int)
print(type(height)) #(float)
print(type(is_student)) #(bool)


# String (str)
city = "Ilorin"
quote = "Python is fun"

name = "John"
print(name.upper())
print(name.lower())
print(len(name))
print()

# joining text
first = "John"
last = "Daniel"
print(first + " " + last)
print(f"{first} {last}")
print()

# Integer (int) - whole numbers
age = 25
score = 75
a = 10
b = 3

print(a + b)
print(a - b)
print(a * b)
print(a / b)
print(a // b)
print(a % b)
print()

# float (float) - decimal numbers
price = 10.5
pi = 3.14159


# Boolean (bool) True/False
is_logged_in = True
has_paid = False

age = 20
print(age >= 18)
print(age == 18)
print()

# Comment: notes that python ignores
# This stores the user's age
age = 50

# type casting: converting between types
x = "5"
y = "2"
print(x + y)
print()

# convert text to number
x  = int("5")
y = int("2")
print(x + y)
print()

# convert to float
weight_text = ("72.5")
weight = float("72.5")
print(weight + 1.0)
print()

# covert from number to string
age = 25
print("age: " +str(age))
print()

# example: input()
age_text = input("Enter your age: ")
age = int(age_text)
print(age + 5)
print()

age = int(input("Enter your age:: "))
print(age + 5)
```

Operators

```python
# Arithmetic operators (+, -, *, /, //, **, %)
# + addition
2 + 3 # 5

# also work in joining (concatenating) text
"Py" + "thon"

# - subtraction
10 - 4 # 6

# * multiplication
6 * 7  # 42
# string repetition
text = "ha" * 3
print(text)
print()

# / division (always gives a float)
10 / 2 # 5.0
9 / 2 # 4.5

# // floor division (rounds down to the nearest whole number)
10 // 3 # 3

# ** exponentiation
2 ** 3 # 8
5 ** 2 # 25

# % modulus (remainder)
10 % 3 # 1 (3*3=9, remainder 1)
# check odd/even numbers
n = 11
n % 2 == 1 # True

n = 10
n % 2 == 1 # False

# order of operation
# Python follows the normal math rule 1. Parenthesis(), 2. Power**, 3. Multiplication/Division * / // %, 4. Addtion/Subtraction + -
2 + 3 * 4
(2 + 3) * 4

# = assignment operator
score = 10
name = "Tunde"
```
